```python
tuscartas=[]
cartasbot= []

carta= ""
posicioncarta= ""
fichas= 1000
apuesta= 0
turno= 0
```

```python
#AQUI AGARRA UNA CARTA ALEATORIA
def cartaleatoria(cartas):
    numeroaleatorio = random.randint(0, len(cards) -1)
    return cartas[numeroaleatorio]
## Faltaría quitar la carta seleccionada del deck (para que no vuelva a salir?)
```

```python
#AQUI IMPRIME LAS CARTAS QUE SALGAN Y LES PONE UN VALOR
def repartidor():
    carta= cartaleatoria(cards)
    print(carta)
    posicioncarta= cards.index(carta) + 1
    if posicioncarta > 10:
            posicioncarta= 10
    if turno == 0:   
        if posicioncarta == 1:
            if sum(cartasbot) < 10:
                posicioncarta= 11
            else: 
                posicioncarta= 1
    elif turno== 1:
        if posicioncarta == 1:
            if sum(tuscartas) < 10:
                posicioncarta= 11
            else: 
                posicioncarta= 1
    return posicioncarta
```

```python
#AQUI PREGUNTA SI QUIERES UNA CARTA ADICIONAL
def otracarta():
    print ('¿Quieres otra carta (Si o No)')
    return input().lower().startswith('s')
```

```python
#AQUI PREGUNTA SI QUIERES JUGAR DE NUEVO
def jufardenuevo():
    print ('¿Quieres jugardenuevo (Si o No)')
    return input().lower().startswith('s')
```

```python
#AQUI PREGUNTA SI QUIERE DOBLAR LA APUESTA
def jugardenuevo():
    print ('¿Quieres doblar la apuesta? (Si o No)')
    return input().lower().startswith('s')
```

```python
#AQUI SE ESTABLECE QUIEN GANA O QUIEN PIERDE
def reglas():
    global fichas
    if sum(tuscartas) > 21:
        print("PIERDES")
        print("⛃ " + str(fichas))
        if fichas == 0:
            print("te quedaste sin fichas, gracias por jugar")
            return False
        else:
            if jugardenuevo():
                return fichas
            else:
                return False
            
    else:
        while sum(cartasbot) >= 17:
            cartasbot.append(repartidor())
        print(sum(cartasbot))
        if sum(cartasbot) > 21:
            print("GANASTE")
            fichas= fichas + (apuesta*2)
            print("⛃ " + str(fichas))
            if jugardenuevo():
                return fichas
            else:
                return False
        
        elif sum(cartasbot) <= 21:
            if sum(cartasbot) > sum(tuscartas):
                print("PIERDES")
                print("⛃ " + str(fichas))
                if fichas == 0:
                    print("te quedaste sin fichas, gracias por jugar")
                    return False
                else:
                    if jugardenuevo():
                        return fichas
                    else:
                        return False
                
            elif sum(cartasbot) <= sum(tuscartas):
                print("GANASTE")
                fichas= fichas + (apuesta*2)
                print("⛃ " + str(fichas))
                if jugardenuevo():
                    return fichas
                else:
                    return False
            else:
                print("EMPATE")
                fichas= fichas + apuesta
                print("⛃ " + str(fichas))
                if jugardenuevo():
                    return fichas
```
